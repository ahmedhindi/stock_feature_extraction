```python
import matplotlib.pyplot as plt
from timaker import TiMaker
import pandas as pd
import numpy as np
import stockstats
from stockstats import StockDataFrame as Sdf
plt.style.use('ggplot')




class IndicatorPlotter:
    def __init__(self, path, to_plot=[]):
        self.path = path
        self.df = None
        self.features = None
        self.to_plot = to_plot

    def get_df(self):
        df = TiMaker(self.path)
        df.run_all()
        self.features = df.features
        self.to_plot = self.features.keys()
        self.df = df.final_data


    def contains(self,subset):
        return [col for col in self.df.columns if subset in col]

    def plot_all(self):
        figsize = (20,10)
        list_of_tis = list(self.features.keys())
        list_of_tis = list(map(str.lower, list_of_tis))
        f,  ax = plt.subplots(len(list_of_tis), sharex=True)
        for i, ti in enumerate(list_of_tis):
            if ti == 'bbands':
                tp = self.df[self.contains(ti)]
                tp.plot(ax=ax[i], figsize=(15,10), legend=False, title=ti)
                self.df.close.plot(ax=ax[i], legend=False, title=ti)
            else:
                tp = self.df[self.contains(ti)]
                tp.plot(ax=ax[i], figsize=figsize, legend=False, title=ti)    
```

```python
sample = IndicatorPlotter('data.csv')
sample.get_df()
sample.to_plot = 'macd'
sample.df = sample.df[sample.df.columns.unique()]
```

```python
sample.plot_all()
sample.features
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(4*np.pi*x)
d = pd.DataFrame([x,y1,y2]).T
```

```python
d.columns = ['x','y1', 'y2']
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)

ax1.fill_between(x, 0, y1)
ax1.set_ylabel('between y1 and 0')

ax2.fill_between(x, y1, 1)
ax2.set_ylabel('between y1 and 1')

ax3.fill_between(x, y1, y2)
ax3.set_ylabel('between y1 and y2')
ax3.set_xlabel('x')
```

```python
plt.plot(d.x,d.y1)
plt.fill_between(d.x,0, d.y1)
```
