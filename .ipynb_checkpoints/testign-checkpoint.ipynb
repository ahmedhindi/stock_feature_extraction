```python
import datetime as datetime
import matplotlib.ticker as ticker
```

```python
def plotter(data, pattern_name):
    fig, ax = plt.subplots(figsize=(32,12))
    candlestick2_ohlc(ax,data['Open'],data['High'],data['Low'],data['Close'],width=0.6) 
    plt.vlines(data.index[data[pattern_name]],data.Close.min(), data.Close.max())
    plt.title(pattern_name, fontsize=25)
```

```python
d = data.data
d.index = np.arange(len(d))
d=d[:-200]
for pattern_name in d.columns[6:]:
    plotter(d,pattern_name=pattern_name)
```
